```python
import pandas as pd
import locale
import pywhatkit as kit
locale.setlocale(locale.LC_ALL, 'es_ES.utf-8')
file = pd.ExcelFile('G:\Mi unidad\IARAFcloud\Datos Fiscal\FISCAL\Transferencias Nacionales a Provincias\Coparticipación Análisis.xlsx')
df=file.parse('Hoja5')
meses = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']
mes_id = int(input('Mes: '))
mes = meses[mes_id-1]
mes1 = meses[mes_id]
anio = int(input('Año: '))
infla_mes = float(input('Inflación del mes: '))
infla_ia = float(input('Inflación interanual: '))
texto = ['','','','']
for i in range(4):
    if (df.loc[i,'Var Copa R']>0):
        var_r = 'aumento'
        var_r_w = 'creció'
        perd_text = 'el aumento'
        perd_text_w = 'menos'
    else:
        var_r = 'caída'
        var_r_w = 'cayó'
        perd_text = 'la caída'
        perd_text_w = 'más'
    if(df.loc[i,'Var Acum']>0):
        var_acum = 'un crecimiento'
    else:
        var_acum = 'una disminución'
    if(df.loc[i,'Perd Acum']>0):
        perd_acum = 'La pérdida real acumulada'
        var_acum_r_w ='pérdida'
    else:
        perd_acum = 'El aumento real acumulado'
        var_acum_r_w = 'crecimiento'
    perd_porc = df.loc[i,'Perd Acum']/df.loc[i,'Copa promedio']
    texto[i] = '*Proyecciones de coparticipación a la Provincia de ' + df.loc[i,'Provincia'] + ' para el resto del año*\n \nEn el mes de ' + mes + ', *la provincia recibió $' + str(locale.format_string('%d',df.loc[i,'Copa'],1)) + ' millones*, mostrando una variación del ' + str(round(df.loc[i,'Var Copa']*100,1)) + '% con respecto a ' + mes + ' de ' + str(anio-1) + ' (' + var_r + ' del ' + str(round(df.loc[i,'Var Copa R']*100,1)) + '% en términos reales, que equivale a $' + str(locale.format_string('%d',abs(df.loc[i,'Perdida']),1)) + ' millones en pesos de ' + mes + ' ' +str(anio) + ' ).\n \n*En lo que va del año, la provincia recibió $' + str(locale.format_string('%d',df.loc[i,'Acum'],1)) + ' millones*. Esto implica ' + var_acum + ' nominal del ' + str(round(df.loc[i,'Var Acum']*100,1)) + '% (' + str(round(df.loc[i,'Var Acum R']*100,1)) + '% en términos reales). ' + perd_acum + ' en el año equivale a $' + str(locale.format_string('%d',df.loc[i,'Perd Acum'],1)) + ' millones, es decir, _en los últimos ' + str(mes_id) + ' meses la pérdida acumulada equivale a un ' + str(abs(round(perd_porc*100,1))) + '% de una coparticipación mensual típica_.\n\n*Proyecciones para el resto del año* \n\nIaraf mantiene tres escenarios alternativos para la coparticipación, pero recomienda tomar decisiones en base al pesimista. Así, la coparticipación hacia ' + df.loc[i,'Provincia'] + ' muestra tres caminos posibles:\n\n*a) Base sin COVID*: se proyectan los meses restantes del año suponiendo una normalización de la recaudación, tomando como base la dinámica observada en los meses previos a la pandemia. En lo que queda del año, *recibiría $' + str(locale.format_string('%d',df.loc[i,'Scovid Faltantes'],1)) + ' millones*, lo que implicaría que finalice el año con un crecimiento nominal del ' + str(round(df.loc[i,'Scovid Var']*100,1)) + '%. \n*b) COVID Optimista*: se supone que la recaudación seguirá deprimida por efectos de la pandemia, volviendo de a poco a la nueva normalidad, hasta alcanzarla en el mes de ' + mes1 + ', por lo que ahora _es idéntico al Base sin COVID_. \n*c) COVID Pesimista*: se supone que no habrá un retorno a la normalidad en este año, por lo que los efectos de la pandemia perdurarían hasta diciembre. En lo que queda del año, *recibiría $' + str(locale.format_string('%d',df.loc[i,'Pes Faltantes'],1)) + ' millones*, lo que implicaría que finalice el año con un crecimiento nominal del ' + str(round(df.loc[i,'Pes Var']*100,1)) + '%.\nEn los 3 escenarios, la inflación promedio esperada para el período es del ' + str(round(infla_ia*100,1)) + '% (vs. mismo período ' + str(anio-1) + ').\n\nCada escenario puede mensualizarse, suponiendo las dinámicas de recuperación explícitas de cada uno. Así, el escenario Base sin COVID muestra una recuperación inmediata a partir de ' + mes1 + ', al igual que el COVID Optimista. Por otro lado, el COVID Pesimista no muestra una vuelta a la normalidad este año. De este proceso surgen los escenarios mensuales que pueden verse en el cuadro.\n\n'
    texto_word[i] = 'Proyecciones de coparticipación a la Provincia de ' + df.loc[i,'Provincia'] + ' para el resto del año\n\nEn el mes de ' + mes + ', la provincia recibió $' + str(locale.format_string('%d',df.loc[i,'Copa'],1)) + ' millones, mostrando una variación del ' + str(round(df.loc[i,'Var Copa']*100,1)) + '% con respecto a ' + mes + ' de ' + str(anio-1) + '. Cuando se considera la inflación del período, y suponiendo una tasa de aumento del IPC para ' + mes + ' de ' + str(round(infla_mes*100,1)) + '%, la coparticipación ' + var_r_w + ' un ' + str(round(df.loc[i,'Var Copa R']*100,1)) + '%, en términos reales.\n\nPara dimensionar ' + perd_text + ' de la coparticipación, se puede expresar en cuánta plata tendría que haber recibido la provincia para, considerando la inflación, no perder ni ganar recursos. Para mantener el poder adquisitivo constante, deberían haberse recibido $' + str(locale.format_string('%d',abs(df.loc[i,'Perdida']),1)) + ' millones ' + perd_text_w + ' (en pesos de ' + mes + ' ' + str(anio) + ').\n\nConsiderando lo que va del año, la provincia recibió $' + str(locale.format_string('%d',df.loc[i,'Acum'],1)) + ' millones. Esto implica ' + var_acum + ' nominal del ' + str(round(df.loc[i,'Var Acum']*100,1)) + '% (' + var_acum_r_w + ' de ' + str(round(df.loc[i,'Var Acum R']*100,1)) + '% en términos reales). ' + perd_acum + ' en el año equivale a $' + str(locale.format_string('%d',df.loc[i,'Perd Acum'],1)) + ' millones, es decir, en los últimos ' + str(mes_id) + ' meses la pérdida acumulada equivale a un ' + str(abs(round(perd_porc*100,1))) + '% de una coparticipación mensual típica.\n\nProyecciones para el resto del año\n\nComo les envié el mes pasado, Iaraf mantiene tres escenarios alternativos que consideran diferentes dinámicas posibles para la evolución de la pandemia, las consiguientes medidas de aislamiento, y sus efectos en la actividad económica y recaudación nacional:\na.	Base sin COVID: se proyectan los meses restantes del año suponiendo una normalización de la recaudación, tomando como base la dinámica observada en los meses previos a la pandemia. En lo que queda del año, recibiría $' + str(locale.format_string('%d',df.loc[i,'Scovid Faltantes'],1)) + ' millones, lo que implicaría que finalice el año con un crecimiento nominal del ' + str(round(df.loc[i,'Scovid Var']*100,1)) + '%. \nb.	COVID Optimista: se supone que la recaudación seguirá deprimida por efectos de la pandemia, volviendo de a poco a la nueva normalidad, hasta alcanzarla en el mes de ' + mes1 + ', por lo que ahora es idéntico al Base sin COVID. \nc.	COVID Pesimista: se supone que no habrá un retorno a la normalidad en este año, por lo que los efectos de la pandemia perdurarían hasta diciembre. En lo que queda del año, recibiría $' + str(locale.format_string('%d',df.loc[i,'Pes Faltantes'],1)) + ' millones, lo que implicaría que finalice el año con un crecimiento nominal del ' + str(round(df.loc[i,'Pes Var']*100,1)) + '%.\nEn los 3 escenarios, la inflación promedio esperada para el período es del ' + str(round(infla_ia*100,1)) + '% (vs. mismo período ' + str(anio-1) + ').\n\nCada escenario puede mensualizarse, suponiendo las dinámicas de recuperación explícitas de cada uno. Así, el escenario Base sin COVID muestra una recuperación inmediata a partir de ' + mes1 + ', al igual que el COVID Optimista. Por otro lado, el COVID Pesimista no muestra una vuelta a la normalidad este año. De este proceso surgen los escenarios mensuales que pueden verse en el cuadro.\n\n'
print(texto[0] + texto[1] + texto[2]+ texto[3])
print(texto_word[0] + texto_word[1] + texto_word[2]+ texto_word[3])
#kit.sendwhatmsg("+5493513037702", texto[0], 14, 51)
#kit.sendwhatmsg("+5493513037702", texto[1], 14, 52)
#kit.sendwhatmsg("+5493513037702", texto[2], 14, 53)
#kit.sendwhatmsg("+5493513037702", texto[3], 14, 54)
```
